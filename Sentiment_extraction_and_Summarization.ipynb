```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import numpy as np
import nltk
nltk.download('punkt')
```

```python
data=pd.read_csv('/content/drive/MyDrive/Assignment.csv', names=['Article'],skiprows=[0])
```

```python
df=pd.DataFrame(data)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sentiment = SentimentIntensityAnalyzer()
```

```python
sentiments=[]
for i in range(0,len(df)):
  s = re.sub(r'[^\w\s]','',df['Article'][i])
  s = s.replace('\n',' ')
  s= s.replace('Reuters','')
  x=sentiment.polarity_scores(s)
  sentiments.append(list(x.values())[:-1])
```

```python
sentiments
```

```python
len(sentiments)
```

```python
scores=[]
np.array(sentiments)
for i in sentiments:
  if np.argmax(i)==1:
    scores.append('Neutral')
  elif np.argmax(i)==0:
    scores.append('Negative')
  else :scores.append('Positive')
scores

```

```python
len(scores)
```

```python
np.argsort(sentiments[1])[1]
```

```python
tilt=[]
for i in range(0,len(sentiments)):
  if np.argsort(sentiments[i])[1]==1:
    tilt.append('Neutral')
  elif np.argsort(sentiments[i])[1]==0:
    tilt.append('Negative')
  else :tilt.append('Positive')
tilt
```

```python
df_sentiments=pd.DataFrame({'Primary sentiment':scores, 'Sentiment tilting towards':tilt})
df_sentiments
```

```python
df_sentiments=pd.concat([df,df_sentiments],axis=1)
```

```python
df_sentiments
```

```python
df_sentiments.to_csv('sentiments.csv')
```

```python
from transformers import LongformerTokenizer, EncoderDecoderModel
model = EncoderDecoderModel.from_pretrained("patrickvonplaten/longformer2roberta-cnn_dailymail-fp16")
tokenizer = LongformerTokenizer.from_pretrained("allenai/longformer-base-4096")
```

```python
len(df['Article'][0])
```

```python
summaries=[]
for i in range(len(df)):
  article = df['Article'][i]
  if len(nltk.word_tokenize(article))>2800:
    article1=article[:len(article)//2]
    input_ids = tokenizer(article1, return_tensors="pt").input_ids
    output_ids = model.generate(input_ids)
    sum1=tokenizer.decode(output_ids[0], skip_special_tokens=True)


    article2=article[len(article)//2:]
    input_ids = tokenizer(article2, return_tensors="pt").input_ids
    output_ids = model.generate(input_ids)
    sum2=tokenizer.decode(output_ids[0], skip_special_tokens=True)

    summaries.append(sum1+sum2)


  else:
    input_ids = tokenizer(article, return_tensors="pt").input_ids
    output_ids = model.generate(input_ids)
    summaries.append(tokenizer.decode(output_ids[0], skip_special_tokens=True))
```

```python
summaries=pd.DataFrame(summaries)
```

```python
summaries.to_csv('summaries.csv')
```
